###  Importing all the required libraries

```python
import numpy as np
import pandas as  pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
time_format = '%Y%m%d-%H:%M:%S:%f'
df['Time'] = pd.to_datetime(df['Time'], format=time_format)
# df['Date'] = df['Time'].dt.date
df['Time'] = df['Time'].apply(lambda x: x.strftime('%Y%m%d-%H:%M'))
```

### Tokenization and Word Frequemcy Analysis

```python
from sklearn.feature_extraction.text import CountVectorizer

# Tokenize EventTemplate
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(temp_df['EventTemplate'])

# Get word frequencies
word_frequencies = pd.DataFrame(X.toarray(), columns=vectorizer.get_feature_names_out())

# Plot the top N most frequent words
top_words = word_frequencies.sum().sort_values(ascending=False).head(10)
plt.figure(figsize=(10, 6))
top_words.plot(kind='bar', color='skyblue')
plt.title('Top 10 Most Frequent Words in EventTemplate')
plt.xlabel('Words')
plt.ylabel('Frequency')
plt.show()
```

### EVent Template Length Distribution

```python
# Add a column for the length of EventTemplate
df['EventTemplateLength'] = df['EventTemplate'].apply(len)

# Plot the distribution of EventTemplate lengths
plt.figure(figsize=(10, 6))
sns.histplot(df['EventTemplateLength'], bins=20, kde=True, color='orange')
plt.title('Distribution of EventTemplate Lengths')
plt.xlabel('EventTemplate Length')
plt.ylabel('Frequency')
plt.show()
```
